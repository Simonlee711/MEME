# Scratch Notebook

```python
import pandas as pd
import os
import glob
```

```python
path = "./mimic-iv-clinical-database-demo-2.2/hosp/"
csv_files = glob.glob(os.path.join(path, "*.csv.gz"))
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    print()
```
